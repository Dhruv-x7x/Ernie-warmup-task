```python
# CONVERT EACH PAGE OF PDF TO AN IMAGE FOR OCR
from pdf2image import convert_from_path

pages = convert_from_path("/content/test.pdf", dpi=300)

for i, page in enumerate(pages):
    page.save(f"page_{i}.png", "PNG")
```

```python
# DO OCR ON EACH PAGE
ocr = PaddleOCR(
    use_doc_orientation_classify=False,
    use_doc_unwarping=False,
    use_textline_orientation=False,
    lang='en',
)
results = []
for i, page in enumerate(pages):
    print(f"Doing page number {i}")
    img_path = f"/content/page_{i}.png"
    page.save(img_path, "PNG")

    ocr_result = ocr.predict(img_path)
    results.append(ocr_result)
```

```python
structured = []

for page_idx, page in enumerate(results):
    page_blocks = []
    for block in page:
        for text, score in zip(block["rec_texts"], block["rec_scores"]):
            page_blocks.append({
                "text": text,
                "confidence": float(score)
            })
    structured.append({
        "page": page_idx,
        "content": page_blocks
    })
```

```python
with open("document.json", "w") as f:
    json.dump(structured, f, indent=2)
```

```python
def ocr_to_markdown(pages):
    md = []
    for page in pages:
        md.append(f"\n\n## Page {page['page']+1}\n")
        for block in page["content"]:
            md.append(block["text"])
    return "\n".join(md)

markdown_text = ocr_to_markdown(structured)

with open("document.md", "w") as f:
    f.write(markdown_text)
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=access_token,
    base_url="https://aistudio.baidu.com/llm/lmapi/v3",
)

system_prompt = """
You are a professional web developer.
Convert the provided markdown into a single, clean HTML file after summarizing every paragraph.
Use semantic HTML5, headings, paragraphs, and fancy embedded CSS.
No explanations. Output ONLY valid HTML.
"""

response = client.chat.completions.create(
    model="ernie-5.0-thinking-preview",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": markdown_text}
    ],
    max_tokens=10000,
    extra_body={"web_search": {"enable": False}}
)
```

```python
html = response.choices[0].message.content
```

```python
with open("index.html", "w") as f:
    f.write(html)
```
